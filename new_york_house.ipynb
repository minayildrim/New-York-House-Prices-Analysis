```python
import pandas as pd
nyc_hp = pd.read_csv("NY-House-Dataset.csv")
nyc_hp.head()
```

```python
nyc_hp.info()
```

```python
nyc_hp.describe()
```

```python
mean_nyc=nyc_hp["PRICE"].mean()
std_nyc=nyc_hp["PRICE"].std()
max_nyc=nyc_hp["PRICE"].max()
mode_nyc=nyc_hp["PRICE"].mode()
median_nyc=nyc_hp["PRICE"].median()
#statistical calculations using pandas
```

```python
print("The mean of house prices:",mean_nyc)
print("The standart deviation of house prices:",std_nyc)
print("The maximum value of house prices:", max_nyc)
print("The mode of house prices:", mode_nyc)
print("The median of house prices:", median_nyc)
```

```python
#cumulative statistics
cumsum_nyc=nyc_hp["PRICE"].cumsum()
cummin_nyc=nyc_hp["PRICE"].cummin()
cummax_nyc=nyc_hp["PRICE"].cummax()
print("cumulative sum of house prices",cumsum_nyc)
print("cumulative min of house prices",cummin_nyc)
print("cumulative max of house prices",cummax_nyc)
```

```python
unique_houses= nyc_hp.drop_duplicates(subset=["MAIN_ADDRESS","STATE"])
print(unique_houses) #to drop the values which are identical, it is done to make sure that all houses are listed just only once.
```

```python
unique_houses["STREET_NAME"].value_counts(sort=True)
#it shows where the houses are located based on street names, sorted by ascending order
```

```python
unique_houses["STREET_NAME"].value_counts(normalize=True)
#We see the ratio, based on street names
#20% of houses are located in Manhattan.
```

```python
price_man=unique_houses[unique_houses["STREET_NAME"]=="Manhattan"]["PRICE"].mean()
price_brook=unique_houses[unique_houses["STREET_NAME"]=="Brooklyn"]["PRICE"].mean()
price_qu=unique_houses[unique_houses["STREET_NAME"]=="Queens"]["PRICE"].mean()
price_stat=unique_houses[unique_houses["STREET_NAME"]=="Staten Island"]["PRICE"].mean()
price_bronx=unique_houses[unique_houses["STREET_NAME"]=="The Bronx"]["PRICE"].mean()

#Getting mean of top ranked streets
```

```python
print("The mean price of Manhattan is ", price_man)
print("The mean price of Brooklyn is ", price_brook)
print("The mean price of Queens is ", price_qu)
print("The mean price of Staten Island is ", price_stat)
print("The mean price of The Bronx is ", price_bronx)
#it is suitable for less data, therefore it should be done by the help of "groupby"
```

```python
unique_houses.groupby("STREET_NAME")["PRICE"].mean()
#the mean of houses prices of each street
```

```python
unique_houses.groupby("STREET_NAME")["PRICE"].agg([min,max,sum])
#to get more statistical values in one line
```

```python
unique_houses.groupby(["STREET_NAME","TYPE"])["PRICE"].mean()
```

```python
import numpy as np
unique_houses.pivot_table(values="PRICE", index="STREET_NAME", aggfunc=np.median)
#getting medians of each street name based on the price
```

```python
import numpy as np
unique_houses.pivot_table(values="PRICE", index="STREET_NAME", aggfunc=[np.median, np.mean])
#multiple statistics, getting medians and means of each street name based on the price
```

```python
sorted_houses=unique_houses.sort_index()
print(sorted_houses) #sorting
```

```python
data=unique_houses.loc[:,"PRICE": "STREET_NAME"]
print(data)
#to get values from price to street name, street name is not included
```

```python
print(unique_houses.loc[unique_houses["PRICE"]>900000])
#to see how many houses are over 900000
```

```python
print(unique_houses.iloc[0:3])
#slicing for first 3 values
```

```python
print(unique_houses.iloc[3,2])
#getting the price of second row by using indexing
```

```python
price_street=unique_houses.pivot_table("PRICE", index=["STREET_NAME"], columns=["TYPE"])
print(price_street)
#doing pivot table to see prices, indexing by street name with the type columns
```

```python
price_street.loc[("Brooklyn: Dumbo"):("Manhattan: Queens")] #slicing in pivot tables
```

```python
price_street.loc[("Brooklyn: Dumbo"):("Manhattan: Queens"),"Coming Soon":"House for sale"]
#slicing
```

```python
mean_price=price_street.mean()
```

```python
# Filter for the street that had the highest mean price
print(mean_price[mean_price == mean_price.max()])
```

```python
# Filter for the street that had the lowest mean price
print(mean_price[mean_price == mean_price.min()])
```

```python
#visualization of the data by matplotlib
import matplotlib.pyplot as plt
```

```python
#showing the mean price, related to street
graph_data = unique_houses.groupby("STREET_NAME")["PRICE"].mean()
graph_data = graph_data.sort_values(ascending=False)
graph_data.head()
```

```python
graph_data.plot(kind="bar", title="House Prices in New York, Based on Streets")
plt.show()
#bar chart

#since it impossible to see every street in the graph, i will visualize 10 the most expensive streets on the bar chart
```

```python
graph_10=graph_data.head(10)
graph_10=graph_data.head(10).plot(kind="bar",title="The Most Expensive House Prices in New York, Based on Streets")
plt.show()
#bar chart
#10 the most expensive streets on the bar chart
```

```python
graph2 = unique_houses.groupby("TYPE")["PRICE"].mean()
graph2.plot(kind="bar", title= "Average House Prices Based on Type")
plt.show()
```

```python
graph2 = unique_houses.groupby("TYPE")["PRICE"].std()
graph2.plot(kind="bar", title= "Standart Deviation House Prices Based on Type")
plt.show()
#this shows us the standart deviation by types. Standart deviation shows us the spread around the mean


```
